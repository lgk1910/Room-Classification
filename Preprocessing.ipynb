```python
BATHROOM_PATH = 'dataset/Bathroom/'
```

```python
bathroom_subfolders = os.listdir(LIVING_ROOM_PATH)
```

```python
bathroom_files = []
for subfolder in bathroom_subfolders:
    bathroom_files.append(os.listdir(BATHROOM_PATH+subfolder))

bathroom_files = np.array(bathroom_files)
bathroom_files.shape
for i in range(0,4):
    print(np.array(bathroom_files[i]).shape)
```

```python
print(bathroom_files[3][0])
```

```python
bathroom_samples = []
count = 0
for i in range(4):
    for file in bathroom_files[i]:
#         print(BATHROOM_PATH + living_room_subfolders[i] + "/" +file)
        img = cv2.imread(BATHROOM_PATH + bathroom_subfolders[i] + "/" +file)
        try:
          resized_img = cv2.resize(img,(150,150))
          bathroom_samples.append(resized_img)
        except:
          pass
        if count%100==0:
          print(f'Successfully process picture {count}th')
        count += 1
        # plt.matshow(resized_img)
        # break
```

```python
bathroom_samples = np.array(bathroom_samples)
bathroom_samples.shape
```

```python
bathroom_labels = np.zeros(bathroom_samples.shape[0])
bathroom_labels.shape
```
